```python
import pandas as pd
df=pd.read_csv('/content/weather_data.csv',parse_dates=["day"])
df.set_index('day',inplace=True)
df
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({'temperature':0,'windspeed':0,'event':'no event'})
new_df
```

```python
new_df=df.fillna(method='ffill')
new_df
```

```python
new_df=df.fillna(method="bfill",axis="columns")
new_df
```

```python
new_df=df.fillna(method="ffill",limit=2)
new_df
```

```python
new_df=df.interpolate()
new_df
```

```python
new_df=df.interpolate(method="time")
new_df
```

```python
new_df=df.dropna()
new_df
```

```python
new_df=df.dropna(how="all")
new_df
```

```python
new_df=df.dropna(thresh=2)
new_df
```
